```python
from langchain.document_loaders import PyPDFLoader

loaders = PyPDFLoader("/content/drive/MyDrive/Generative-AI-and-LLMs-for-Dummies.pdf",
                     "/content/drive/MyDrive/Data Mining Processing & Transformation.pdf"


                     )
docs = []
for loader in loaders:
    docs.extend(loader.load())
```
